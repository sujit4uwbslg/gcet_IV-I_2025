1. Normalization of Input Data

Normalization is used to scale input values to a smaller range, typically 
[0,1] or [−1,1].

Min-Max normalization scales a value \( x \) to a range \([0, 1]\) using:

$$
x' = \frac{x - x_{min}}{x_{max} - x_{min}}
$$

where:
- \( x \) = original value  
- \( x_{min} \) = minimum value in the dataset  
- \( x_{max} \) = maximum value in the dataset  
- \( x' \) = normalized value

```python
import numpy as np

# Sample data
X = np.array([2, 4, 6, 800, 10], dtype=float)

# Min-Max Normalization
X_min = np.min(X)
X_max = np.max(X)
X_normalized = (X - X_min) / (X_max - X_min)

print("Normalized Input:", X_normalized)
```


Other possible normalization functions are given in the following table:


| Normalization Method       | Formula                                           | Use Case / When to Use                             |
|---------------------------|---------------------------------------------------|----------------------------------------------------|
| Z-Score Standardization   | $x' = \frac{x - \mu}{\sigma}$                     | Normalize to zero mean and unit variance           |
| Max Abs Scaling           | $x' = \frac{x}{\max(|x|)}$                        | Preserves sparsity; range $[-1, 1]$                |
| Robust Scaling            | $x' = \frac{x - \text{median}}{\text{IQR}}$       | Robust to outliers                                 |
| L2 Normalization          | $x' = \frac{x}{\|x\|_2} = \frac{x}{\sqrt{x_1^2 + x_2^2 + \cdots + x_n^2}}$ | Used for cosine similarity, word vectors  |
| L1 Normalization          | $x' = \frac{x}{\|x\|_1} = \frac{x}{|x_1| + |x_2| + \cdots + |x_n|}$       | Sparse models and text normalization               |
| Decimal Scaling           | $x' = \frac{x}{10^j}$ where $j$ is smallest s.t. $|x'| < 1$ | Simple scaling, older datasets          |
| Log Transformation        | $x' = \log(x + 1)$                                | Reduces skewness in right-tailed data              |
| Box-Cox Transformation    | $x' = \frac{x^\lambda - 1}{\lambda},\ \lambda \ne 0$ <br> $x' = \log(x),\ \lambda = 0$ | Only for $x > 0$ |
| Yeo-Johnson Transformation| Handles $x \le 0$ and $x > 0$ (see full definition) | More general than Box-Cox                          |
| Unit Vector Scaling       | $x_i' = \frac{x_i}{\sqrt{\sum_{j=1}^{n} x_j^2}}$   | Normalize each feature vector to unit length       |


What is an Activation Function?

An activation function is a mathematical operation applied to the output of a neuron (node) in a neural network. It determines whether that neuron should be activated (fire) or not, and how much signal to pass on to the next layer.

Why is it needed?

Without activation functions, a neural network would simply be a linear model, no matter how many layers you add. Activation functions introduce non-linearity, allowing the network to learn complex patterns, such
- Images

- Speech

- Text

- Nonlinear decision boundaries

How does it work?

Each neuron performs:
$$
    f(x)=w \cdot x + b
$$
Then applies:
$$
𝑎=g(𝑧)
$$

Where:

- 𝑥: input

- 𝑤: weight

- 𝑏: bias

- 𝑧: linear transformation

- g(𝑧): activation function (e.g., ReLU, Sigmoid)







Here's a list of all the basic activation functions used in neural networks, along with their formulas and use-cases.

1. **Sigmoid:**

The sigmoid activation function maps any real-valued number to a value between 0 and 1.

$$
\sigma(x) = \frac{1}{1 + e^{-x}}
$$

where:

- \( x \): Input to the neuron (weighted sum)
- \( e \): Euler's number (approx. 2.718)
- $\sigma(x)$: Output of the the sigmoid function (activation)

```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid function definition
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generate input values
x = np.linspace(-20, 20, 100)
y = sigmoid(x)

# Plot
plt.figure(figsize=(6, 4))
plt.plot(x, y, label='Sigmoid Function')
plt.title('Sigmoid Activation Function')
plt.xlabel('x')
plt.ylabel('σ(x)')
plt.grid(True)
```

  2. **Tanh (Hyperbolic Tangent):**

The tanh activation function maps any real-valued number to a value between -1 and 1.

$$
\tanh(x) = \frac{e^{x} - e^{-x}}{e^{x} + e^{-x}}
$$

where:

- \( x \): Input to the neuron (weighted sum)
- \( e \): Euler's number (approx. 2.718)
- \( $\tanh(x)$ \): Output of the tanh function (activation)

```python
import numpy as np
import matplotlib.pyplot as plt

# Tanh function definition
def tanh(x):
    return np.tanh(x)

# Generate input values
x = np.linspace(-10, 10, 100)
y = tanh(x)

# Plot
plt.figure(figsize=(6, 4))
plt.plot(x, y, label='Tanh Function', color='orange')
plt.title('Tanh Activation Function')
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.grid(True)
plt.legend()
```

3. **ReLU (Rectified Linear Unit):**

The ReLU activation function outputs the input directly if it is positive; otherwise, it outputs zero.

$$
\mathrm{ReLU}(x) = \max(0, x)
$$

where:

- \( x \): Input to the neuron (weighted sum)
- \( $\mathrm{ReLU}(x)$ \): Output of the ReLU

```python
import numpy as np
import matplotlib.pyplot as plt

# ReLU function definition
def relu(x):
    return np.maximum(0, x)

# Generate input values
x = np.linspace(-10, 10, 100)
y = relu(x)

# Plot
plt.figure(figsize=(6, 4))
plt.plot(x, y, label='ReLU Function', color='green')
plt.title('ReLU Activation Function')
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.grid(True)
plt.legend()
plt.show()
```

4. **Leaky ReLU (Leaky Rectified Linear Unit):**

The Leaky ReLU activation function allows a small, non-zero gradient when the input is negative.

$$
\mathrm{LeakyReLU}(x) = 
\begin{cases}
x & \text{if } x \geq 0 \\
\alpha x & \text{if } x < 0
\end{cases}
$$

where:

- \( x \): Input to the neuron (weighted sum)
- \( \alpha \): Slope for negative input (typically a small value like 0.01)
- \( $\mathrm{LeakyReLU}(x)$ \): Output of the Leaky ReLU function

```python
import numpy as np
import matplotlib.pyplot as plt

# Leaky ReLU function definition
def leaky_relu(x, alpha=0.01):
    return np.where(x >= 0, x, alpha * x)

# Generate input values
x = np.linspace(-10, 10, 100)
y = leaky_relu(x)

# Plot
plt.figure(figsize=(6, 4))
plt.plot(x, y, label='Leaky ReLU Function', color='purple')
plt.title('Leaky ReLU Activation Function')
plt.xlabel('x')
plt.ylabel('LeakyReLU(x)')
plt.grid(True)
plt.legend()
plt.show()
```

5. **Exponential Linear Unit (ELU):**

The ELU activation function helps to avoid the problem of vanishing gradients and allows negative values when the input is less than zero.

$$
\mathrm{ELU}(x) = 
\begin{cases}
x & \text{if } x \geq 0 \\
\alpha (e^{x} - 1) & \text{if } x < 0
\end{cases}
$$

where:

- \( x \): Input to the neuron (weighted sum)
- \( $\alpha$ \): Slope for negative input (a positive constant, typically 1.0)
- \( $\mathrm{ELU}(x)$ \): Output of the ELU function (activation)

```python
import numpy as np
import matplotlib.pyplot as plt

# ELU function definition
def elu(x, alpha=1.0):
    return np.where(x >= 0, x, alpha * (np.exp(x) - 1))

# Generate input values
x = np.linspace(-10, 10, 100)
y = elu(x)

# Plot
plt.figure(figsize=(6, 4))
plt.plot(x, y, label='ELU Function', color='brown')
plt.title('Exponential Linear Unit (ELU) Activation Function')
plt.xlabel('x')
plt.ylabel('ELU(x)')
plt.grid(True)
plt.legend()
plt.show()
```

6. **Swish:**

The Swish activation function is a smooth, non-monotonic function that can outperform ReLU on deeper models.

$$
\mathrm{Swish}(x) = x \cdot \sigma(\beta x) = \frac{x}{1 + e^{-\beta x}}
$$

where:

- \( x \): Input to the neuron (weighted sum)
- \( $\sigma(\beta x)$ \): Sigmoid function applied to \( \beta x \)
- \( $\beta $\): Trainable or fixed parameter (commonly set to 1)
- \( $\mathrm{Swish}(x) $\): Output of the Swish function (activation)

```python
import numpy as np
import matplotlib.pyplot as plt

# Swish function definition
def swish(x, beta=1):
    return x / (1 + np.exp(-beta * x))

# Generate input values
x = np.linspace(-10, 10, 100)
y = swish(x)

# Plot
plt.figure(figsize=(6, 4))
plt.plot(x, y, label='Swish Function', color='blue')
plt.title('Swish Activation Function')
plt.xlabel('x')
plt.ylabel('Swish(x)')
plt.grid(True)
```

**7. Softmax Activation Function**

The softmax function converts a vector of values into a probability distribution, where the sum of all probabilities is 1.

$$
\mathrm{softmax}(z_i) = \frac{e^{z_i}}{\sum_{j=1}^{K} e^{z_j}}
$$

where:

- \($ z_i $\): The input value (logit) for the \(i\)-th class
- \( $K $\): Total number of classes
- \($ e $\): Euler's number (approx. 2.718)
- \( $\mathrm{softmax}(z_i) $\): Output probability for the \(i\)-

```python
import numpy as np
import matplotlib.pyplot as plt

# Softmax function definition
def softmax(z):
    exp_z = np.exp(z - np.max(z))  # for numerical stability
    return exp_z / np.sum(exp_z)

# Generate input values (vector)
z = np.linspace(-2, 2, 5)
y = softmax(z)

# Plot
plt.figure(figsize=(6, 4))
plt.stem(range(len(z)), y)  # Removed use_line_collection
plt.title('Softmax Activation Function')
plt.xlabel('Class Index')
plt.ylabel('Probability')
```

**8. Gaussian Error Linear Unit (GELU):**

The GELU activation function is a smooth, non-linear function that combines properties of both ReLU and sigmoid. It is commonly used in modern neural networks, especially in transformer architectures.

$$
\mathrm{GELU}(x) = x \cdot \Phi(x)
$$

where:

- \( x \): Input to the neuron (weighted sum)
- \( $\Phi(x)$ \): The cumulative distribution function (CDF) of the standard normal distribution

An approximate and commonly used form is:

$$
\mathrm{GELU}(x) \approx 0.5x \left[1 + \tanh\left(\sqrt{\frac{2}{\pi}} \left(x + 0.044715x^3\right)\right)\right]
$$

where:

- \( $\tanh$ \): Hyperbolic tangent function
- \( $\sqrt{\frac{2}{\pi}}$ \): Scaling constant
- \( $0.044715$ \): Empirical constant for approximation
- \( $\mathrm{GELU}(x) $\): Output of the GELU function

```python
import numpy as np
import matplotlib.pyplot as plt

# GELU function definition (approximate version)
def gelu(x):
    return 0.5 * x * (1 + np.tanh(np.sqrt(2/np.pi) * (x + 0.044715 * x**3)))

# Generate input values
x = np.linspace(-10, 10, 100)
y = gelu(x)

# Plot
plt.figure(figsize=(6, 4))
plt.plot(x, y, label='GELU Function', color='red')
plt.title('Gaussian Error Linear Unit (GELU) Activation Function')
plt.xlabel('x')
plt.ylabel('GELU(x)')
plt.grid(True)
```
